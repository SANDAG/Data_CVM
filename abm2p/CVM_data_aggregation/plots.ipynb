```python
import pandas as pd
```

```python
actorType_industry_map = {
    'Industry': 'Industrial',
    'Retail': 'Retail',
    'Transport': 'Transport',
    'Wholesale': 'Wholesale',
    'Service': 'Service/Gov/Office/FA',
    r'Government\Office': 'Service/Gov/Office/FA',
    'Fleet Allocator': 'Service/Gov/Office/FA',
}

industry_category = pd.CategoricalDtype(
    categories=[
        'Industrial',
        'Retail',
        'Service/Gov/Office/FA',
        'Transport',
        'Wholesale',
    ],
    ordered=True,
)
```

```python
tod_map = {
    1: 'OE (3am - 6am)',
    2: 'AM (6am - 9am)',
    3: 'MD (9am - 3:30pm)',
    4: 'PM (3:30pm - 7pm)',
    5: 'OL (7pm - 3am)',
}

tod_category = pd.CategoricalDtype(
    categories=[
        'OE (3am - 6am)',
        'AM (6am - 9am)',
        'MD (9am - 3:30pm)',
        'PM (3:30pm - 7pm)',
        'OL (7pm - 3am)',
    ],
    ordered=True,
)
```

```python
mode_light_map = {
    'Drive Alone': 'Light',
    'Light Heavy Duty Truck': 'Other',
    'Medium Heavy Duty Truck': 'Other',
    'Heavy Heavy Duty Truck': 'Other',
}
```

```python
industry_category
```

```python
tours = (
    pd.read_csv('T:/data/base_2022/CVM/Calib_CVM_B/model_output/commercialVehicleTours.csv')
    .assign(industry=lambda df: df.actorType.map(actorType_industry_map).astype(industry_category))
    .assign(depart_tod=lambda df: df.departTimeFiveTod.map(tod_map).astype(tod_category))
    .assign(mode_light=lambda df: df.tourMode.map(mode_light_map))
    .set_index('tourID')
)
tours.head(20)
```

```python
tours.tourMode.value_counts()

# ? Drive Alone is light or other?
```

```python
trips = (
    pd.read_csv('T:/data/base_2022/CVM/Calib_CVM_B/model_output/commercialVehicleTrips.csv')
    .set_index(['tripID', 'tourID', 'stopID'])
)
trips
```

```python
tours = (
    tours
    .merge(
        trips.reset_index().groupby('tourID')['weightTrip'].max(),
        on='tourID'
    )
)
```

```python
trips = (
    trips
    .merge(
        tours[['tourMode', 'industry']],
        on='tourID'
    )
)
```

```python
# Table 17: Target Tour start Time Shares
# Count

(
    tours
    .reset_index()
    .groupby(['industry', 'depart_tod'])
    ['tourID']
    .count()
    .reset_index()
    .pivot(
        index='industry',
        columns='depart_tod',
    )
    [[('tourID', x) for x in tod_category.categories.to_list()]]
    .apply(
        lambda x: x / (
            tours
            .reset_index()
            .groupby('industry')
            ['tourID']
            .count()
        ) 
    )
)
```

```python
# Table 17: Target Tour start Time Shares
# & Figure 9a-9b Tour Start Time Calibration Results (Industry)
# Naive Weight (max trip weight in tour)

(
    tours
    .reset_index()
    .groupby(['industry', 'depart_tod'])
    ['weightTrip']
    .sum()
    .reset_index()
    .pivot(
        index='industry',
        columns='depart_tod',
    )
    [[('weightTrip', x) for x in tod_category.categories.to_list()]]
    .apply(
        lambda x: x / (
            tours
            .reset_index()
            .groupby('industry')
            ['weightTrip']
            .sum()
        ) 
    )
).loc[:, 'weightTrip'].to_csv('T:/data/base_2022/CVM/Calib_CVM_B/model_table_17_figure9.csv')
```

```python
# Figure 10 Vehicle Type Calibration Results
# Count

# Aggregating by tours (should do by trips/weighted trips?)
# Note, this is considering "Drive Alone" as "Light"

(
    tours
    .reset_index()
    .groupby('industry')
    ['mode_light']
    .value_counts(normalize=True)
    .rename('tours')
    .to_frame()
).sort_index().to_csv('T:/data/base_2022/CVM/Calib_CVM_B/model_figure10.csv')
```

```python
# Figure 12 Trip Length Calibration Results
x = (
    trips
    [['tourMode', 'industry', 'weightTrip', 'distanceTotal']]
    .assign(
        vehClass_indName=(
            lambda df: df.apply(
                lambda row: {
                    ('Drive Alone', 'Industrial'): 'DI',
                    ('Drive Alone', 'Retail'): 'DR',
                    ('Drive Alone', 'Service/Gov/Office/FA'): 'DS',
                    ('Drive Alone', 'Transport'): 'DT',
                    ('Drive Alone', 'Wholesale'): 'DW',
                    ('Light Heavy Duty Truck', 'Industrial'): 'LI',
                    ('Light Heavy Duty Truck', 'Retail'): 'LR',
                    ('Light Heavy Duty Truck', 'Service/Gov/Office/FA'): 'LS',
                    ('Light Heavy Duty Truck', 'Transport'): 'LT',
                    ('Light Heavy Duty Truck', 'Wholesale'): 'LW',                
                    ('Medium Heavy Duty Truck', 'Industrial'): 'MI',
                    ('Medium Heavy Duty Truck', 'Retail'): 'MR',
                    ('Medium Heavy Duty Truck', 'Service/Gov/Office/FA'): 'MS',
                    ('Medium Heavy Duty Truck', 'Transport'): 'MT',
                    ('Medium Heavy Duty Truck', 'Wholesale'): 'MW',                
                    ('Heavy Heavy Duty Truck', 'Industrial'): 'HI',
                    ('Heavy Heavy Duty Truck', 'Retail'): 'HR',
                    ('Heavy Heavy Duty Truck', 'Service/Gov/Office/FA'): 'HS',
                    ('Heavy Heavy Duty Truck', 'Transport'): 'HT',
                    ('Heavy Heavy Duty Truck', 'Wholesale'): 'HW',
                }.get((row.tourMode, row.industry)),
                axis=1
            )
            .astype(
                pd.CategoricalDtype(
                    [
                        'DI',
                        'DR',
                        'DS',
                        'DT',
                        'DW',
                        'LI',
                        'LR',
                        'LS',
                        'LT',
                        'LW',
                        'MI',
                        'MR',
                        'MS',
                        'MT',
                        'MW',
                        'HI',
                        'HR',
                        'HS',
                        'HT',
                        'HW',
                    ],
                    ordered=True,
                )
            )
        )
    )
        .assign(
            weighted_distance=lambda df:(
                df['distanceTotal'] * df['weightTrip']
            )
        )
)
x
```

```python
y = ((x.groupby('vehClass_indName')['weighted_distance'].sum()) / x.groupby('vehClass_indName')['weightTrip'].sum()).rename('model_distance')
y.to_csv('T:/data/base_2022/CVM/Calib_CVM_B/model_figure12.csv')
```
